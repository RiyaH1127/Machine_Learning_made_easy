```python
titanic.head()
```

```python
sns.barplot(x='Pclass',y='Age',hue='Pclass',palette=['red','blue','green'],data=titanic)
#barplot give average average (mean) of numerical column (Age) for each category (Pclass). 
```

```python
sns.barplot(x='Pclass',y='Fare',hue='Sex',data=titanic)
```

```python
#Boxplot(numerical-categorical)
#to see outliers
```

```python
sns.boxplot(x='Sex',y='Age',hue='Survived',data=titanic)
```

```python
#Displot(numerical-categorical)
```

```python
sns.distplot(titanic[titanic['Survived']==0]['Age'])
```

```python
#distplot is deprecated by seaborn
#Use histplot() or kdeplot() ✅
# Histogram -Uses bars,Shows counts in bins.
#KDE-Smooth curve,Shows probability density.
```

```python
sns.histplot(data=titanic, x='Age', hue='Survived', kde=True)
#This shows:
#Bars → histogram
#Curve → KDE line
```

```python
sns.kdeplot(data=titanic, x='Age', hue='Survived', fill=True)
#only smooth curves
#kde- kernel Density Estimation
```

```python
sns.histplot(data=titanic[titanic['Survived'] == 0], x='Age', kde=True)
#not survived
sns.histplot(data=titanic[titanic['Survived'] == 1], x='Age', kde=True)
#survived
```

```python
sns.kdeplot(data=titanic[titanic['Survived'] == 0], x='Age', fill=True)
#probability of not surviving
sns.kdeplot(data=titanic[titanic['Survived'] == 1], x='Age' ,fill=True)
#probability of surviving
```

```python
#HeatMap(categorical-categorical)
#info betn two categorical data
```

```python
pd.crosstab(titanic['Pclass'],titanic['Survived'])
#crosstab- relation betn two categorical data columns
```

```python
sns.heatmap(pd.crosstab(titanic['Pclass'],titanic['Survived']))
#heatmap- shows relationship using hues /colors for values
```

```python
titanic.groupby('Pclass')['Survived'].mean() * 100 #percentage of survival in each class
```

```python
(titanic.groupby('Pclass')['Survived'].mean() * 100).plot(kind='bar')
```

```python
(titanic.groupby('Sex')['Survived'].mean() * 100).plot(kind='bar')
```

```python
(titanic.groupby('Embarked')['Survived'].mean() * 100).plot(kind='bar')
```

```python
#clustermap(categorical-categorical)
```

```python
sns.clustermap(pd.crosstab(titanic['Parch'],titanic['Survived']))
#clustermap-closeness betn values- almost same hues
```
